```python
import pandas as pd
import pandas_profiling
import re
import string
from textblob import TextBlob
import matplotlib.pyplot as plt
```

```python
news_headlines = pd.read_csv('../../data/news_headlines/news_headline_data.csv')
```

```python
news_headlines.head()
```

```python
news_headlines.profile_report()
```

```python
pol = lambda x: TextBlob(x).sentiment.polarity
sub = lambda x: TextBlob(x).sentiment.subjectivity

news_headlines['polarity'] = news_headlines['article_title'].apply(pol)
news_headlines['subjectivity'] = news_headlines['article_title'].apply(sub)
```

```python
publisher_stats = []
for x in news_headlines['publisher'].unique():
    p = news_headlines[news_headlines['publisher'] == x]['polarity']
    s = news_headlines[news_headlines['publisher'] == x]['subjectivity']
    publisher_stats.append([x,p.mean(),s.mean()])
```

```python
publisher_stats
```

```python
fig, axs = plt.subplots(1)

for i in publisher_stats:
    x = i[1]
    y = i[2]
    axs.scatter(x, y, color='#F92672')
    axs.text(x+.003, y+.001, i[0], fontsize=10, color = '#AE81FF')
    axs.tick_params(axis='x', colors='#E6DB74')
    axs.tick_params(axis='y', colors='#E6DB74')
    axs.set_facecolor('#272822')
    axs.set_xlabel('<-- Negative -------- Positive -->', fontsize=12, color = '#AE81FF')
    axs.set_ylabel('<-- Facts -------- Opinions -->', fontsize=12, color = '#AE81FF')
```

```python
def clean_text_round1(text):
    '''Make text lowercase, remove punctuation, and remove words containing numbers.'''
    text = text.lower()
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text

round1 = lambda x: clean_text_round1(x)
```

```python
news_headlines['clean_headline'] = news_headlines['article_title'].apply(round1)
```

```python
#for x in news_headlines['publisher'].unique():
headline_by_publisher = news_headlines.groupby('publisher')['clean_headline'].apply(' '.join).reset_index()
headline_by_publisher.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english')
data_cv = cv.fit_transform(headline_by_publisher.clean_headline)
data_dtm = pd.DataFrame(data_cv.toarray(), columns = cv.get_feature_names())
data_dtm.index = headline_by_publisher.publisher
data_dtm
```

```python
from wordcloud import WordCloud
from sklearn.feature_extraction import text
# Add new stop words
stop_words = text.ENGLISH_STOP_WORDS.union(['vaccine','vaccines','amid', 'michael'])
wc = WordCloud(stopwords=stop_words, background_color="white", colormap="Dark2",
               max_font_size=150, random_state=42)
```

```python
plt.rcParams['figure.figsize'] = [16, 6]
index = 0
# Create subplots for each comedian
for i in headline_by_publisher['publisher']:
    headline = headline_by_publisher[headline_by_publisher['publisher'] == i]['clean_headline']
    #print(headline[0])
    wc.generate(headline[index])
    plt.subplot(3, 4, index+1)
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.title(i, fontsize=12, color = '#AE81FF')
    index = index + 1
    
plt.show()
```
